## Setup
I use two datasets for this project: my own personal streaming history that you can request from Spotify and a Spotify Features dataset that I got from Kaggle. (https://www.kaggle.com/zaheenhamidani/ultimate-spotify-tracks-db)




```python
my_data = pd.read_csv('StreamingHistory.csv')
features = pd.read_csv('SpotifyFeatures.csv')
```

## Cleanup

```python
# Only Feature that I want is Valence
features.drop(["popularity"],1,inplace=True)
features.drop(["track_id"],1,inplace=True)
features.drop(["time_signature"],1,inplace=True)
features.drop(["instrumentalness"],1,inplace=True)
features.drop(["liveness"],1,inplace=True)
features.drop(["acousticness"],1,inplace=True)
features.drop(["key"],1,inplace=True)
features.drop(["mode"],1,inplace=True)
features.drop(["tempo"],1,inplace=True)
features.drop(["loudness"],1,inplace=True)
features.drop(["speechiness"],1,inplace=True)
features.drop(["artist_name"],1,inplace=True)
features.drop(["genre"],1,inplace=True)
features.drop(["danceability"],1,inplace=True)
features.drop(["duration_ms"],1,inplace=True)
features.drop(["energy"],1,inplace=True)
```

```python
features.head()
```

```python
# Removing duplicates from features dataframe. Each song had a row for each genre. 
features = features.drop_duplicates(['track_name'], keep='last')
down = features[features['track_name']=="Down Girl"] 
down
```

## Merging Features and MyData 

```python
features = features.rename(columns={'track_name': 'trackName'}) #Renaming track_name to trackName
df3 = pd.merge(my_data,features,on="trackName")
df3.head()
```

## Sort data

```python
df3 = df3.sort_values(by=['Unnamed: 0']) #Sorting dataframe by endTime 
df3.head()
```

## Splitting date and time 

```python
# Creating a easier way to search by date
# Instead of having to add the time of the day, we can just search the date. 
temp = []
for i in df3["endTime"]:
    temp.append(i.split())

date = []
for j in temp:
    date.append(j[0])
df3["date"] = date 
df3.head()
```

## Changing msPlayed to minsPlayed

```python
df3['msPlayed'] = df3['msPlayed'].divide(60000) #Change milliseconds to minutes 
df3 = df3.rename(columns={'msPlayed': 'minsPlayed'}) #rename msPlayed to minsPlayed
```

## Using Ipywidgets (MUCH better search function)
The Ipywidgets library enabled me to create an interactive search function. Now, I do not have to type out the dates. I can simply just scroll up and down the chart and click. 
Morever, we can search the title of a song to see if we played it during that day. As you type, the chart refreshes live based on the letter that has been typed.

```python
def search_layout(db,track,date):
    my_data = db
    for i in track.split():
        my_data = my_data[my_data["trackName"].str.contains(i,case=False)]
    my_date = my_data[my_data["date"]==date]
    return my_date
```

```python
interact(lambda Track, Date: search_layout(df3,Track,Date), #Ipywidget function 
        Track = '',
        Date = list(df3['date'].unique()))
```

## Stat of the Day Function
This is where our SpotifyFeatures dataset comes in handy. According to Spotify Developer's site, valence "describes the musical positiveness conveyed by a track." 
 - For example, tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry).

Using this, we can determine the mood of the day purely based on the songs listened to.
Most Played Song, Most Played Artist, and Listening can all be calculated from my_data dataset. 

```python
import matplotlib as mpl
from matplotlib import pyplot as plt


def day_stats(db,date):
    my_data = df3
    
    # Most Played Song
    my_song = my_data[my_data["date"]==date]
    most_played = my_song["trackName"].value_counts()[:1].index.tolist() # Gets the song title of the most played song
    my_song = my_song[my_song["trackName"]==most_played[0]] # Returns dataframe with only that song's attributes
    most_played.append(my_song["artistName"].unique()[0]) # Adds artist name to the most_played array 
    print(most_played[0] + " by " + most_played[1] +" is your most played song of the day") 
      
    # Most Played Artist 
    my_artist = my_data[my_data["date"]==date]
    most_artist = my_song["artistName"].value_counts()[:1].index.tolist() # Gets most played artist
    print("Your Most Played Artist of the Day is " + most_artist[0])

    # Listening Time 
    my_time = my_data[my_data["date"]==date] 
    total_time = my_time["minsPlayed"].sum() # Sums up the minsPlayed column for that date
    print("You have listened to a total of " + str(round(total_time)) + " minutes today!")
    
    # Mood of the Day
    my_mood = my_data[my_data["date"]==date]
    #mood = my_mood["valence"].mean() #If we want the average of all the valences of the day instead of the most played song's valence
    mood = my_mood[my_mood["trackName"]==most_played[0]]["valence"].unique()[0] # Gets the valence of the most played song
    
    cmap = plt.cm.get_cmap('RdYlGn') # We use a RdYlGn colormap so that Red is valence of 0 and Green to indicate valence 1. 
    # The redder the color the more negative the song, and vice versa for green. 
    color = mpl.colors.to_hex(list(cmap(mood))[:3]) # Converts (r,g,b,a) output of the cmap() function to hexadecimal
    # Plotting the square
    plt.figure(figsize=(4,4))
    ax = plt.axes()
    # Setting the background color
    ax.set_facecolor(color) # Fills the square with the color that represents the valence or mood
    ax.axes.get_xaxis().set_visible(False) # Getting rid of x-axis 
    ax.axes.get_yaxis().set_visible(False) # Getting rid of y-axis
    plt.title("Mood of the Day")
    
```

```python
interact(lambda Date: day_stats(df3,Date),
        Date = list(df3['date'].unique()))
```
